```python
class User:
    def __init__(self, user_id, name):
        self.user_id = user_id
        self.name = name
```

```python
class Message:
    def __init__(self, sender, receiver, text):
        self.sender = sender
        self.receiver = receiver
        self.text = text
```

```python
class NotificationService:
    def notify(self, user, message):
        print(f"Notification to {user.name}: New message received")
```

```python
class ChatService:
    def __init__(self):
        self.messages = []             
        self.notifier = NotificationService()

    def send_message(self, message):
        self.messages.append(message)
        self.deliver_message(message)

    def deliver_message(self, message):
        print(f"{message.sender.name} -> {message.receiver.name}: {message.text}")
        self.notifier.notify(message.receiver, message)

    def show_chat_history(self):
        print("\nChat History:")
        for msg in self.messages:
            print(f"{msg.sender.name} -> {msg.receiver.name}: {msg.text}")
```

```python
user1 = User(1, "Mithun")
user2 = User(2, "Praveen")
```

```python
chat = ChatService()
```

```python
msg1 = Message(user1, user2, "Hi !")
chat.send_message(msg1)
```

```python
msg2 = Message(user2, user1, "Hello !")
chat.send_message(msg2)
```

```python
chat.show_chat_history()
```
